```python
from sklearn import linear_model

clean = spotify_youtube_data.dropna(subset = ['Stream','Views'])




x = np.reshape(clean[['Stream']], [len(clean[['Stream']]),1])

y = clean[['Views']]
reg = linear_model.LinearRegression()

reg.fit(x,y)
#reg.coef_

```

```python
reg.intercept_
```

```python


sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8000000000), ylim=(0, 8000000000))

plt.xlabel('Spotify Streams')
plt.ylabel('Youtube Views')
plt.title('Youtube Views Compared to Spotify Streams on the Same Song')
plt.plot(x, reg.predict(x),color='k')
plt.show()
```
